```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
dataset=pd.read_csv("Churn_Modelling.csv")
dataset
```

```python
dataset.head(3)
```

```python
# Preprocessing the data
# Drop irrelevant columns

dataset=dataset.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
dataset
```

```python
# Encode Categorical Variables

lg=LabelEncoder()
dataset['Gender']=lg.fit_transform(dataset['Gender'])
dataset
```

```python
# OneHot Encode 'Geography'
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
gohe=ohe.fit_transform(dataset[['Geography']])
gohe
```

```python
gohe.toarray()
```

```python
ohe.get_feature_names_out(['Geography'])
```

```python
jk=pd.DataFrame(gohe.toarray(),columns=ohe.get_feature_names_out(['Geography']))
jk
```

```python
# combine one hot encoder columns with the original data

dataset=pd.concat([dataset.drop('Geography', axis=1), jk], axis=1)
dataset
```

```python
with open('lg.pkl', 'wb') as file:
    pickle.dump(lg, file)

with open('ohe.pkl', 'wb') as file:
    pickle.dump(ohe, file)
```

```python
# Divide the dataset into independant and dependant features

x=dataset.drop('Exited', axis=1)
y=dataset['Exited']


# split the data in training and testing sets

x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=42)

# Scale these features

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
with open ('scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)
```
