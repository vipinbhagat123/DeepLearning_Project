```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
dataset=pd.read_csv("Churn_Modelling.csv")
dataset
```

```python
dataset.head(3)
```

```python
# Preprocessing the data
# Drop irrelevant columns

dataset=dataset.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
dataset
```

```python
# Encode Categorical Variables

lg=LabelEncoder()
dataset['Gender']=lg.fit_transform(dataset['Gender'])
dataset
```

```python
# OneHot Encode 'Geography'
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
gohe=ohe.fit_transform(dataset[['Geography']])
gohe
```

```python
gohe.toarray()
```

```python
ohe.get_feature_names_out(['Geography'])
```

```python
jk=pd.DataFrame(gohe.toarray(),columns=ohe.get_feature_names_out(['Geography']))
jk
```

```python
# combine one hot encoder columns with the original data

dataset=pd.concat([dataset.drop('Geography', axis=1), jk], axis=1)
dataset
```

```python
with open('lg.pkl', 'wb') as file:
    pickle.dump(lg, file)

with open('ohe.pkl', 'wb') as file:
    pickle.dump(ohe, file)
```

```python
# Divide the dataset into independant and dependant features

x=dataset.drop('Exited', axis=1)
y=dataset['Exited']


# split the data in training and testing sets

x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=42)

# Scale these features

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
with open ('scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)
```

##ANN IMPLEMENTATION

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
# Build Our ANN Model

model=Sequential([
    Dense(64, activation='relu', input_shape=(x_train.shape[1],)),   ##HL1 Connected with input layer
    Dense(32, activation='relu'), ## HL2
    Dense(1, activation='sigmoid') # Output Layer
])
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
loss
```

```python
## Compile the model

model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
## Set up the Tensorboard

from tensorflow.keras.callbacks import EarlyStopping, TensorBoard

log_dir='logs/fit' + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
# Set up Early Stopping

early_stopping_calllback = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
# Train the model

history=model.fit(
    x_train, y_train, validation_data=(x_test, y_test), epochs=100,
    callbacks=[tensorflow_callback, early_stopping_calllback]
)
```

```python
model.save('model.h5')
```
